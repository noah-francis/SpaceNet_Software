```python
# heres what i do to plot things also 
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (10, 4), dpi = 200)
ax = fig.add_subplot(1, 1, 1)
ax.grid()
ax.set_ylabel('cos(x) [~]')
ax.set_xlabel('x [rad]')
ax.set_title('Title Here')
ax.set_xlim(-20,20)
ax.set_ylim(-1.2, 1.2)

x = arange(-10*pi, 10*pi, 0.1)
y = cos(x)

# ax.plot(x, y)
ax.scatter(x, y, s=1, c='b', label='cosine')
ax.legend(loc=1)
plt.show()
```
